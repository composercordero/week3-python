```python
### Ingredient must match measurement
class Cocktails:
    def __init__(self, name, ingredients, instructions, glass):
        self.name = name
        self.ingredients = ingredients
        self.instructions = instructions
        self.glass = glass
        
    def __str__(self):
        output = f"\n----{self.name}----\n"
        output += f"\nGlass type: {self.glass}\n"
        output += f"\nInstructions: {self.instructions}\n"
        output += f"\nIngredients:"
        for i,m in self.ingredients.items():
            output += f"\n\t{i} - {m}"
        return output
        
#     def __repr__(self):
#         return f"< {self.name} | 

class CocktailsAPI:
    
    def __init__(self, key):
        self.base_url = "https://www.thecocktaildb.com/api/json/v1/"
        self.api_key = key
            
    def _get(self, q=''):
        res1 = requests.get(f'{self.base_url}/{self.api_key}/search.php?s={q}')
        if res1.ok:
            return res1.json()
        else:
            return "invalid search, please try again"
            
    def get_cocktail(self, drink_name):
        data = self._get(drink_name)
        ingredients = {}
        if data and data['drinks']: # Check if the dictionary exists and has the data we are looking for
            drink = data['drinks'][0]
            title = drink['strDrink']
            instructions = drink['strInstructions']
            glass = drink['strGlass']  
            for i in range(1,16):
                if drink[f'strIngredient{i}']:
                    ingredients[drink[f'strIngredient{i}']] = drink[f'strMeasure{i}']
            new_cocktail = Cocktails(title, ingredients, instructions, glass)  
            return new_cocktail
        else:
            return f"{drink_name.title()} is not on our list"

user1 = CocktailsAPI('1')
new = user1.get_cocktail('margarita')

print(new)
```
